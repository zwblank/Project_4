```python
# Import the required modules
import pandas as pd
from pathlib import Path
```

```python
# Read the usage_stats.csv file from the Resources folder into a Pandas DataFrame
realestate_df = pd.read_csv(
    Path("Resources/realtor-data.zip.csv")
)

# Review the DataFrame
display(realestate_df)
```

```python
columns_to_check = ['status', 'price', 'bed', 'bath', 'acre_lot', 'city', 'state', 'zip_code', 'house_size', 'prev_sold_date']
filtered_realestate_df = realestate_df
filtered_realestate_df.dropna(subset=columns_to_check, inplace=True)

filtered_realestate_df
```

```python
columns_to_drop = ['brokered_by', 'street', 'prev_sold_date']
dropped_columns_realestate_df = filtered_realestate_df.drop(columns=columns_to_drop, axis=1)

dropped_columns_realestate_df
```

```python
clean_realestate_df = dropped_columns_realestate_df[~dropped_columns_realestate_df['state'].isin(['Puerto Rico', 'Virgin Islands', 'Guam', 'New Brunswick', 'Alaska', 'Hawaii'])]

clean_realestate_df
```

```python
missing_value_counts =clean_realestate_df.isna().sum()
print("Missing value counts per column:")
print(missing_value_counts)
```

```python
clean_realestate_df.to_csv('Resources/clean_realestate_data.csv', index=False)
```
